```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
from pylab import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 6, 6
warnings.filterwarnings("ignore")
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
sns.distplot(df.total_bill, kde=True)
```

```python
sns.boxplot(df.total_bill)
```

```python
ax = sns.violinplot(x=df.total_bill, inner="box")
```

```python
sns.distplot(df[(df.day == 'Fri') & (df.smoker == 'Yes')]['total_bill'])
```

```python
sns.distplot(df[(df.day == 'Fri') & (df.smoker == 'No')]['total_bill'])
```

```python
sns.boxplot(df[(df.day == 'Fri')]['total_bill'])
```

```python
ax = sns.violinplot(x=df.day, y=df.total_bill, hue=df.smoker,
                    data=df, palette="muted", split=True)
```
